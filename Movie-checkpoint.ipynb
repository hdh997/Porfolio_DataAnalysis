```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the size of the plot

pd.set_option('display.max_rows',None)

df = pd.read_csv('C:/Users/admin/Desktop/Practice_data/Movie/movies.csv')
```

```python
df.head()
```

# Cleaning data

```python
#Check for missing data
print(df.isnull().sum())
```

```python
df_w_null = df[df.isnull().any(axis= 1)]
print(df_w_null)
```

```python
#replace missing data
for col in df.columns:
    if df[col].isnull().any() and df[col].dtypes == 'object':
        df[col] = df[col].fillna('N/A')
    elif df[col].isnull().any() and df[col].dtypes == 'float64':
        df[col] = df[col].fillna(0)
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df.head(20)
```

```python
df.tail(20)
```

```python
#Change data type
df['budget']=df['budget'].astype(int)
df['gross']=df['gross'].astype(int)
```

```python
#Sort df by gross col

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

# Create correlation btween

    - Company vs gross
    - budget vs gross


```python
#scatter plot budget w gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross Earning')
plt.ylabel('Budget For Film')
plt.show()
```

```python
#heatmap for corr
corr_matrix=df.corr()
sns.heatmap(df.corr(),annot=True)
plt.title('Correlation Matrix for Numeric Columns')
plt.show()
```

# Convert All non numeric cols to numeric type

```python
numeric_df = df
for col in numeric_df:
    if numeric_df[col].dtype == 'object':
        numeric_df[col] = numeric_df[col].astype('category')
        numeric_df[col] = numeric_df[col].cat.codes
numeric_df.head(20)
```

```python
corr_matrix=numeric_df.corr()
sns.heatmap(numeric_df.corr(),annot=True)
plt.title('Correlation Matrix for All Columns')
plt.show()
```

```python
corr_mat = numeric_df.corr()
corr_pairs = corr_mat.unstack()
sorted_pairs =corr_pairs.sort_values()
sorted_pairs
```

```python
#Only high correlation pairs
high_corr = sorted_pairs[(sorted_pairs) > 0.5 ]
high_corr
```
